```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
# Load digits dataset
digits = load_digits()
X = digits.data
y = digits.target
```

```python
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
```

```python
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
# Plot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='tab10', ax=ax[0], legend='full')
ax[0].set_title('PCA (2D)')
ax[0].axis('off')

sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y, palette='tab10', ax=ax[1], legend='full')
ax[1].set_title('t-SNE (2D)')
ax[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y, palette='tab10', ax=ax[1], legend='full')
ax[1].set_title('t-SNE (2D)')
ax[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
tsne = TSNE(n_components=2,
            perplexity=50,
            learning_rate=200,
            n_iter=1000,
            random_state=42)

X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y, palette='tab10', legend='full')

plt.axis('off')
plt.tight_layout()
plt.show()
```
