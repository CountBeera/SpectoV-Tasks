```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
X_blobs, y_blobs = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
X_blobs = StandardScaler().fit_transform(X_blobs)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_blobs[:, 0], X_blobs[:, 1])
plt.title("Original make_blobs Dataset (True Labels)")
plt.grid(True)
plt.show()
```

# K-Means


```python
kmeans_blobs = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans_blobs.fit_predict(X_blobs)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=kmeans_labels, cmap='Set1', s=40, edgecolor='k')
plt.scatter(kmeans_blobs.cluster_centers_[:, 0], kmeans_blobs.cluster_centers_[:, 1], s=200, c='yellow', edgecolor='black', marker='X', label='Centroids')
plt.legend()
plt.title("KMEANS")
plt.grid(True)
plt.show()
```

# DBSCAN 

```python
dbscan_blobs = DBSCAN(eps=0.6, min_samples=5)
dlables = dbscan_blobs.fit_predict(X_blobs)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=dlables, cmap='Set1', s=40, edgecolor='k')
plt.title("DBSCAN")
plt.grid(True)
plt.show()
```
